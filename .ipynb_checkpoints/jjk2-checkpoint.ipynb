```python
def lecture_csv_flexible(path):
    if not os.path.exists(path):
        raise FileNotFoundError(path)
    for sep in [',', ';', '\t']:
        try:
            df_verif = pd.read_csv(path, sep=sep, low_memory=False)
           
            if df_verif.shape[1] > 1:
                return df_verif
        except Exception:
            continue
    return pd.read_csv(path, low_memory=False)
```

Charge le fichier CSV et inspecte les colonnes et types. Parse la colonne date .Retourne un DataFrame et affiche un sommaire,d√©tecte colonne date automatiquement si nom diff√©rent

```python
def chargement_dataset(csv_path):
    
    df1 = lecture_csv_flexible(csv_path)
    df2 = df1.copy()
    date_col = None
    df2["value"] = df2["value"].astype(str).str.replace(",", ".")
    df2["value"] = pd.to_numeric(df2["value"], errors="coerce")

    if index in df2.columns:
        date_col = index
    else:
        for c in df2.columns:
            if c.lower() in ('timestamp', 'date', 'datetime', 'time') or 'date' in c.lower() or 'time' in c.lower():
                date_col = c
                print(f"[chargement_dataset] colonne date d√©tect√©e: {date_col}")
                break
    if {'value_type', 'value'}.issubset(set(df2.columns)):
        if date_col is None and 'timestamp' in df2.columns:
            date_col = 'timestamp'
        if date_col is None:
            possible = [c for c in df2.columns if df2[c].dtype == object][:1]
            if possible:
                date_col = possible[0]
        if date_col is None:
            raise ValueError("Aucune colonne de date d√©tect√©e dans le CSV (format long attendu).")
        df2[date_col] = pd.to_datetime(df2[date_col], errors='coerce')
# Ici on  enlve entrees sans date probalement isssue d'erreur d'enregistrement 
        df2 = df2.loc[~df2[date_col].isna()].copy()
        df = df2.pivot_table(index=date_col, columns='value_type', values='value', aggfunc='mean')
        df.index = pd.to_datetime(df.index, errors='coerce')
        df = df.sort_index()
    else:
        if date_col is not None:
            df2[date_col] = pd.to_datetime(df2[date_col], errors='coerce')
            df2 = df2.loc[~df2[date_col].isna()].copy()
            df = df2.set_index(date_col).sort_index()
        else:
            # essayer convertir index en datetime
            try:
                df2.index = pd.to_datetime(df2.index)
                df = df2.sort_index()
            except Exception:
#retourner tel quel et laisser le pipeline √©chouer proprement plus tard  c'est mahleureux mais pas de choix . je suis ouvers a des solution pour gerer des cas pareil 
                df = df2.copy()

    print('Afficher les  premieres  de la copy:')
    try:
        print(df.head())
    except Exception:
        print("Impossible d'afficher l'entete (format inattendu).")
    print('\n Colonnes detectees:', list(df.columns))
    print('\nInfos:')
    print(df.info())
    return df
```

Agr√©ge les mesures multiples par jour en prenant la moyenne.Convertit la colonne date en index datetime,Reechantillonee par jour et calcule la moyenne pour la cible +  les variable explicatives .Retourne df_daily .Pourquoi: l'utilisateur veut une seule valeur moyenne par jour.extraire des colonnes dans notre dataset celle qui nous interesse.
    

```python
def compilation_donnees(df,cible,date_col, colonnes, rule):
  
    df2 = df.copy()
    if not isinstance(df2.index, pd.DatetimeIndex):
        try:
            df2.index = pd.to_datetime(df2.index)
        except Exception:
            raise ValueError("L'index du DataFrame n'est pas convertible en DatetimeIndex dans compilation_donnees.")
#variables explicatives et cibles
    cols = []
    if cible in df2.columns:
        cols.append(cible)
    for c in colonnes:
        if c in df2.columns and c not in cols:
            cols.append(c)
            
    if len(cols) == 0:
        raise ValueError("Aucune colonne cible ou var explicatif pr√©sente pour la compilation des donn√©es.")
    df_compiler = df2[cols].resample(rule).mean()

# Garder un trac√© des nombres de points par jour peut aider pour debug,  on compte sur la colonne cible
    try:
        comptage = df2[cols].resample(rule).count().iloc[:, 0].rename('freq_mesure')
    except Exception:
#comptage des lignes par p√©riode
        comptage = df2.resample(rule).size().rename('freq_mesure')
    df_compiler = df_compiler.assign(counts=comptage)
    return df_compiler
```

G√®re les valeurs manquantes selon la methode dediees.Detecte outliers via IQR sur chaque colonne et remplace par NaN puis interpol .Si la fraction de valeurs manquantes journali√®res est sup√©rieure √† drop_threshold, calcul des Quartiles 1er et 3 e  quartile et  l'iqr l'interquartil range ou encore plage des valeurs normales centrales.

```python
def Valeurs_manquantes_et_aberantes(df_compiler, method='interpolate',seuil_manquant=0.7 ,tolerance=0.2):
    
    data = df_compiler.copy()
    colonnes_numerique = data.select_dtypes(include=[np.number]).columns.tolist()
    for col in colonnes_numerique:
        q1 = data[col].quantile(0.25)
        q3 = data[col].quantile(0.75)
        iqr = q3 - q1
        bornes_inf = q1 - 1.5 * iqr
        bornes_sup = q3 + 1.5 * iqr
#creation de masque de detection
        masque_de_detection = (data[col] < bornes_inf) | (data[col] > bornes_sup)
        if isinstance(masque_de_detection, pd.Series) and masque_de_detection.any():
            print("Des anomalies d√©tect√©es ‚Äî nettoyage en cours...")
            # remplacer outliers par NaN -> permettra interpolation
            data.loc[masque_de_detection, col] = np.nan
        plt.figure(figsize=(12,6))
        
        plt.plot(data.index, data[col], label='Valeur originale', linewidth=2)
        
        if isinstance(masque_de_detection, (pd.Series, np.ndarray)):
            plt.scatter(data.index[masque_de_detection],data.loc[masque_de_detection, Variable_cible],color='red',label='Anomalies d√©tect√©es',s=60,zorder=3)
        
        plt.title("D√©tection des anomalies sur la s√©rie temporelle")
        plt.xlabel("timestamp")
        plt.ylabel(col)
        plt.legend()
        plt.grid()
        plt.tight_layout()
        plt.show()

    frac_manquante = data.isna().mean()
    print('proportions de valeurs manquantes par colonne:')
    print(frac_manquante)
    colonnes_a_supprimer = frac_manquante[frac_manquante > seuil_manquant].index
    data= data.drop(columns=colonnes_a_supprimer)
    print(f"\nColonnes supprim√©es car plus de {seuil_manquant*100}% de valeurs manquantes :")
    print(list(colonnes_a_supprimer))
    print("\n DataFrame apr√®s suppression :")
    print(data.head())
    data=data.drop(columns=['counts'])
    if method == 'ffill':
        data = data.fillna(method='ffill')
    elif method == 'bfill':
        data = data.fillna(method='bfill')
    elif method == 'drop':
        thresh = max(1, int((1 - tolerance) * data.shape[1]))
        data = data.dropna(thresh=thresh)
    else:
        try:
            data = data.interpolate(method='time')
        except Exception:
              raise ValueError("Vous devez choisir une methode pour gerer les valeur manquantes")

    # Si reste encore des NaN en bord, on applique ffill/bfill
    data = data.fillna(method='ffill').fillna(method='bfill')
    return data
```

```python

def test_stationnarite(y, alpha=0.05):
    result = adfuller(y.dropna(), autolag='AIC')
    stat, pval, usedlag, nobs, crit, icbest = result

    print("\n Test de stationnarit√© (ADF)")
    print(f"  - Statistique ADF : {stat:.4f}")
    print(f"  - p-value         : {pval:.4f}")
    print("  - Valeurs critiques :", crit)

    ser_stationary = pval < alpha
    print(f"La s√©rie est {'stationnaire' if ser_stationaire else 'non stationnaire'} (alpha={alpha})")
    return {"adf_stat": stat, "pvalue": pval, "ser_stationaire": ser_stationaire}
```

D√©composition + tests ADF + log + diff√©renciation + visualisation.Sauvegarde automatique des graphiques dans le dossier out_dir.
    

```python

def decomposition_serie_temporelle(df, cible, model='auto', freq=None, plot=True, test_stationaire=True, out_dir='outputs'):
    
    os.makedirs(out_dir, exist_ok=True)
    data = df.copy()

    if cible not in data.columns:
        raise ValueError(f"La colonne cible '{cible}' est absente du DataFrame.")

    # D√©terminer la fr√©quence temporelle
    if freq is None:
        freq = pd.infer_freq(data.index)
        if freq is None:
            freq = 7
        else:
            try:
                freq = pd.Timedelta(freq).days or 7
            except Exception:
                freq = 7

    y = data[cible].astype(float).replace([np.inf, -np.inf], np.nan).interpolate()

    # Test de stationnarit√© initial
    adf_avant = test_stationnarite(y) if test_stationaire else {"ser_stationaire": True, "pvalue": 0.0}

    # Choix automatique du mod√®le
    if model == 'auto':
        ratio = y.std() / np.maximum(y.mean(), 1e-6)
        model_type = 'multiplicative' if ratio > 0.5 else 'additive'
    else:
        model_type = model
    print(f"\nüîç Type de mod√®le saisonnier choisi : {model_type}")

    y_log = None
    y_diff = None
    app_log = False
    diff_app = False

    # Log-transform si n√©cessaire
    if model_type == 'multiplicative' or not adf_apres["ser_stationaire"]:
        print(" S√©rie non stationnaire ou multiplicative --> application d‚Äôun log-transform.")
        y_log = np.log1p(y)
        app_log = True
    else:
        y_log = y.copy()

    # Test apr√®s log
    adf_apres_log = test_stationnarite(y_log, verbose=False)
    if not adf_apres_log["ser_stationary"]:
        print("Toujours non stationnaire apr√®s log --> diff√©renciation appliqu√©e.")
        y_diff = y_log.diff().dropna()
        diff_app = True
    else:
        y_diff = y_log.copy()

    # D√©composition
    try:
        decomposition = seasonal_decompose(y_diff, model='additive', period=int(freq), extrapolate_trend='freq')
        data['tendance'] = decomposition.trend
        data['saisonal'] = decomposition.seasonal
        data['residus'] = decomposition.resid
        data['cible_decom'] = y_diff - decomposition.seasonal
    except Exception as e:
        print(f"[decomposition_serie_temporelle] Erreur d√©composition: {e}")
        data['tendance'] = np.nan
        data['saisonal'] = np.nan
        data['residus'] = np.nan
        data['cible_decom'] = y_diff

    data[['tendance', 'saisonal', 'residus', 'cible_decomp']] = (
        data[['tendance', 'saisonal', 'residus', 'cible_decomp']].interpolate().fillna(method='bfill').fillna(method='ffill'))

    # Visualisation et sauvegarde
    if plot:
        fig, axes = plt.subplots(4, 1, figsize=(12, 10), sharex=True)
        axes[0].plot(y, label="S√©rie brute", color='black')
        axes[0].set_title("1Ô∏è‚É£ S√©rie originale")
        axes[0].legend()

        if app_log:
            axes[1].plot(y_log, label="Apr√®s log-transform", color='blue')
            axes[1].set_title("2Ô∏è‚É£ Apr√®s transformation logarithmique")
            axes[1].legend()
        else:
            axes[1].set_visible(False)

        if diff_app:
            axes[2].plot(y_diff, label="Apr√®s diff√©renciation", color='purple')
            axes[2].set_title("3Ô∏è‚É£ Apr√®s diff√©renciation (stabilisation)")
            axes[2].legend()
        else:
            axes[2].set_visible(False)

        axes[3].plot(data['tendance'], label="Tendance", color='orange')
        axes[3].plot(data['saisonal'], label="Saisonnalit√©", color='green', alpha=0.7)
        axes[3].plot(data['residus'], label="R√©sidus", color='red', alpha=0.6)
        axes[3].set_title("4Ô∏è‚É£ D√©composition finale")
        axes[3].legend()

        plt.tight_layout()

     
        fig_path = os.path.join(out_dir, f"decomposition_{cible}.png")       # Sauvegarde du graphique principal
        plt.savefig(fig_path, dpi=300, bbox_inches='tight')
        plt.show()

        print(f"\n Figure enregistr√©e : {fig_path}")

    resume = {
        "avant_log": adf_avant,
        "apres_log": adf_apres_log,
        "log_app": app_log,
        "differenciation_app": diff_app,
        "model_type": model_type
    }

    return data, resume
```

Cr√©e des features temporelles utilis√©es par les mod√®les ML:
    - Lags de la cible
    - Rolling mean et std
    - Features calendaires: dayofweek, month, day
    - Retourne 
la valeur cible. une moyenne croissante montre une tendance positive et l ecart type eleve montre une serie instable ou bruite

```python
def creation_de_features(data, cible, lags=[], rolling_windows=[]):
    dfc = data.copy()
    if not isinstance(dfc.index, pd.DatetimeIndex):
        try:
            dfc.index = pd.to_datetime(dfc.index)
        except Exception:
            raise ValueError("Index non DatetimeIndex dans creation_de_features")
    if cible not in dfc.columns:
        raise ValueError(f"Cible '{cible}' introuvable dans data pour creation de features.")
    taille = dfc.shape[0]
    # memoire de la variable 
    if taille <= 1000:
        print('Utilise un modele de simple comme la regression lineaire')
        for lag in lags:
            dfc[f'lag_{lag}'] = dfc[cible].shift(lag)
            # Rolling stats
            for w in rolling_windows:
                dfc[f'rol_mean_{w}'] = dfc[cible].shift(1).rolling(window=w, min_periods=1).mean()
                dfc[f'rol_std_{w}'] = dfc[cible].shift(1).rolling(window=w, min_periods=1).std()
            dfc = dfc.interpolate(method='time').fillna(method='ffill').fillna(method='bfill')
        dfc['dayofweek'] = dfc.index.dayofweek
        dfc['month'] = dfc.index.month
        dfc['day'] = dfc.index.day
        try:
            if 'lat' in dfc.columns and 'long' in dfc.columns:
                dfc['lat_lon'] = dfc['lat'] * dfc['long']
            elif 'lat' in dfc.columns and 'lon' in dfc.columns:
                dfc['lat_lon'] = dfc['lat'] * dfc['lon']
        except Exception:
            pass
        
    else:

        for lag in lags:
            dfc[f'lag_{lag}'] = dfc[cible].shift(lag)
            # Rolling stats
            for w in rolling_windows:
                dfc[f'rol_mean_{w}'] = dfc[cible].shift(1).rolling(window=w, min_periods=1).mean()
                dfc[f'rol_std_{w}'] = dfc[cible].shift(1).rolling(window=w, min_periods=1).std()
        dfc['dayofweek'] = dfc.index.dayofweek
        dfc['month'] = dfc.index.month
        dfc['day'] = dfc.index.day
       
        if 'lat' in dfc.columns and 'long' in dfc.columns:
            dfc['lat_lon'] = dfc['lat'] * dfc['long']
        elif 'lat' in dfc.columns and 'lon' in dfc.columns:
            dfc['lat_lon'] = dfc['lat'] * dfc['lon']

    try:
        max_lag = max(lags) if len(lags) > 0 else 0
    except Exception:
        max_lag = 0
    if max_lag > 0:
        dfc = dfc.iloc[max_lag:].copy()
        

    return dfc
```

```python
def Decoupage_train_test_tempo(df, test_size=0.2):
    if len(df) == 0:
        raise ValueError("Le DataFrame est vide avant le d√©coupage ! V√©rifie la cr√©ation des features.")

    n = len(df)
    split = int(n * (1 - test_size))
    train_df = df.iloc[:split].copy()
    test_df = df.iloc[split:].copy()
    return train_df,test_df
```

```python
def Normalisatiion_donnees(train_df, test_df, feature_cols):
    print("Taille avant normalisation :", train_df.shape, test_df.shape)
    if train_df.empty or test_df.empty:
        raise ValueError("Train ou Test vide ‚Äî v√©rifie le d√©coupage temporel ou la cr√©ation des features.")

    scaler = StandardScaler()
    train = train_df.copy()
    test = test_df.copy()
    safe_feature_cols = [c for c in feature_cols if c in train.columns and np.issubdtype(train[c].dtype, np.number)]
    if len(safe_feature_cols) == 0:
        print('IL ya aucune variable dans feature, la creation de ne renvoi rien, verifier le nettoyage')
        return scaler, train, test
    
    train_local = train.copy()
    test_local = test.copy()
    train_local[safe_feature_cols] = train_local[safe_feature_cols].fillna(train_local[safe_feature_cols].mean())
    test_local[safe_feature_cols] = test_local[safe_feature_cols].fillna(train_local[safe_feature_cols].mean())
    train_local[safe_feature_cols] = scaler.fit_transform(train_local[safe_feature_cols])
    test_local[safe_feature_cols] = scaler.transform(test_local[safe_feature_cols])
    return scaler, train_local, test_local
```

```python
def evaluation(y_true, y_pred):
    if isinstance(y_true, pd.DataFrame):
        y_true= y_true.iloc[:, 0]
    y_true = np.array(y_true).reshape(-1)
    y_pred = np.array(y_pred).reshape(-1)
    if len(y_true) != len(y_pred):
        min_len = min(len(y_true), len(y_pred))
        print(f" evaluation: mismatch lengths (y_true={len(y_true)}, y_pred={len(y_pred)}). Truncating to {min_len}.")
        y_true = y_true[:min_len]
        y_pred = y_pred[:min_len]
    mae = mean_absolute_error(y_true, y_pred)
    rmse = np.sqrt(mean_squared_error(y_true, y_pred))
    # √©viter division par z√©ro dans MAPE
    mape = np.mean(np.abs((y_true - y_pred) / np.maximum(np.abs(y_true), 1e-6))) * 100
    return {'MAE': mae, 'RMSE': rmse, 'MAPE (%)': mape}
```

```python
def train_linear_model(model, X_train, y_train, X_test):
    if isinstance(y_train, pd.DataFrame):
        y_train = y_train.iloc[:, 0]
    y_train = np.array(y_train).reshape(-1)
    model.fit(X_train, y_train)
    preds = model.predict(X_test)
    return np.array(preds).reshape(-1)
```

AutoReg de statsmodels n'utilise que y, pas  variable explicative par defaut.On utilise AutoReg pour capter l'auto-correlation directe.

```python
def train_autoreg(X_train, y_train, X_test, lags=7):
    if isinstance(y_train, pd.DataFrame):
        y_train = y_train.iloc[:, 0]
    y_train = pd.Series(np.array(y_train).reshape(-1)).astype(float)
    if len(y_train) <= lags:
        last = float(y_train.iloc[-1]) if len(y_train) > 0 else 0.0
        return np.array([last] * len(X_test)).reshape(-1)
    model = AutoReg(y_train, lags=lags, old_names=False).fit()
    preds = model.predict(start=len(y_train), end=len(y_train) + len(X_test) - 1, dynamic=False)
    return np.array(preds).reshape(-1)
```

```python
def train_sarimax(y_train, X_train, X_test, order=(2,0,2), seasonal_order=(2,0,2,0)):
    if isinstance(y_train, pd.DataFrame):
        y_train = y_train.iloc[:, 0]
    y_train = np.array(y_train).reshape(-1)
    X_train_local = X_train.copy() if hasattr(X_train, 'copy') else pd.DataFrame(X_train)
    X_test_local = X_test.copy() if hasattr(X_test, 'copy') else pd.DataFrame(X_test)
    try:
        model = SARIMAX(endog=y_train, exog=X_train_local, order=order, seasonal_order=seasonal_order,
                        enforce_stationarity=False, enforce_invertibility=False)
        res = model.fit(disp=False)
        preds = res.predict(start=len(y_train), end=len(y_train) + len(X_test_local) - 1, exog=X_test_local)
        return np.array(preds).reshape(-1)
    except Exception as e:
        #renvoyer la derni√®re valeur connue r√©p√©t√© (√©chec d'estimation)
        print(f"[train_sarimax] erreur fit/predict: {e}. Retour constant.")
        last = float(y_train[-1]) if len(y_train) > 0 else 0.0
        return np.array([last] * len(X_test_local)).reshape(-1)
```

```python
def train_random_forest(X_train, y_train, X_test, n_estimators=100):
    if isinstance(y_train, pd.DataFrame):
        y_train = y_train.iloc[:, 0]
    y_train = np.array(y_train).reshape(-1)
    rf = RandomForestRegressor(max_depth=5,min_samples_leaf=5 ,n_estimators=n_estimators, random_state=0)
    rf.fit(X_train, y_train)
    y_preds = rf.predict(X_test)
    return np.array(y_preds).reshape(-1), rf
```

```python
def train_xgboost(X_train, y_train, X_test, params=None, num_round=100):
    if not XGBOOST_AVAILABLE:
        raise ImportError('XGBoost non disponible: pip install xgboost')
    if isinstance(y_train, pd.DataFrame):
        y_train = y_train.iloc[:, 0]
    dtrain = xgb.DMatrix(X_train, label=np.array(y_train).reshape(-1))
    dtest = xgb.DMatrix(X_test)
    if params is None:
        params = {'objective': 'reg:squarederror', 'eval_metric': 'rmse', 'seed': 0}
    bst = xgb.train(params, dtrain, num_round)
    preds = bst.predict(dtest)
    return np.array(preds).reshape(-1), bst
```

```python
def train_lstm(X_train_df, y_train, X_test_df, epochs=50, batch_size=16, units=50):
    X_train = X_train_df.values
    X_test = X_test_df.values
    X_train_3d = X_train.reshape((X_train.shape[0], 1, X_train.shape[1]))
    X_test_3d = X_test.reshape((X_test.shape[0], 1, X_test.shape[1]))

    model = Sequential()
    model.add(LSTM(units, input_shape=(X_train_3d.shape[1], X_train_3d.shape[2])))
    model.add(Dropout(0.2))
    model.add(Dense(1))
    model.compile(optimizer='adam', loss='mse')
    es = EarlyStopping(monitor='loss', patience=5, restore_best_weights=True)
# y_train peut √™tre Series ou array
    if isinstance(y_train, pd.DataFrame):
        y_local = y_train.iloc[:, 0].values.reshape(-1)
    elif isinstance(y_train, pd.Series):
        y_local = y_train.values.reshape(-1)
    else:
        y_local = np.array(y_train).reshape(-1)
# Si trop peu d'√©chantillons
    if len(y_local) < 2:
        raise ValueError("Pas assez d'√©chantillons pour entra√Æner LSTM.")
    model.fit(X_train_3d, y_local, epochs=epochs, batch_size=batch_size, verbose=0, callbacks=[es])
    preds = model.predict(X_test_3d).flatten()
    return np.array(preds).reshape(-1), model
```

Validation temporelle 'expanding window'.
Pour chaque pli:
- entrainer sur[0:split]
- tester sur [split: split+horizon]
Le comportement varie selon model_name; on utilise les wrappers definis ci-dessus.
Retourne liste de metrics par pli et predictions de la derniere fenetre.

```python
def walk_forward_validation(df_finale, colonnes, cible,horizon, model_name='ridge', n_splits=5):
    n = len(df_finale)
    if n <= horizon:
        raise ValueError("Donn√©es trop courtes pour effectuer la validation avec l'horizon demand√©.")
# choisir split points equidistants dans la partie finale
    split_points = split_points = np.unique(np.linspace(int(n*0.6), n - horizon, n_splits, dtype=int))

    all_metrics = []
    derniere_preds =[]
    for i, split in enumerate(split_points):
        end_test = min(split + horizon, len(df_finale))
        train = df_finale.iloc[:split]
        test = df_finale.iloc[split:split+horizon]
# Securit√©: v√©rifier que colonnes existent dans train/test - sinon utiliser intersection
        if isinstance(colonnes, list):
            usable_cols = [c for c in colonnes if c in train.columns]
        else:
# si colonnes est une str (mal pass√©e), convertir
            usable_cols = [colonnes] if colonnes in train.columns else []
        if len(usable_cols) == 0:
            raise ValueError("Aucune colonne feature valide trouv√©e pour l'entra√Ænement dans walk_forward_validation.")
        X_train = train[usable_cols]
        X_test = test[usable_cols]
        if isinstance(cible, list):
            y_train = train[cible[0]]
            y_test = test[cible[0]]
        else:
                y_train = train[cible]
                y_test = test[cible]
        # Forcer 1D numpy
        if isinstance(y_train, pd.DataFrame):
            y_train = y_train.iloc[:, 0]
        if isinstance(y_test, pd.DataFrame):
            y_test = y_test.iloc[:, 0]
        y_train = np.array(y_train).reshape(-1)
        y_test = np.array(y_test).reshape(-1)

        if model_name == 'ridge':
            preds = train_linear_model(Ridge(alpha=1.0), X_train, y_train, X_test)
            model_obj = None
        elif model_name == 'lasso':
            preds = train_linear_model(Lasso(alpha=0.1), X_train, y_train, X_test)
            model_obj = None
        elif model_name == 'ar':
            preds = train_autoreg(X_train, y_train, X_test, lags=7)
            model_obj = None
        elif model_name == 'sarimax':
# utiliser exog = colonnes exog (on suppose que EXOG_COLS sont dans features)
            exog_train = train[[c for c in selection_colonne if c in train.columns]]
            exog_test = test[[c for c in selection_colonne if c in test.columns]]
            preds = train_sarimax(y_train, exog_train, exog_test, order=(2,2,2), seasonal_order=(2,2,2,7))
            model_obj = None
        elif model_name == 'rf':
            preds, model_obj = train_random_forest(X_train, y_train, X_test)
        elif model_name == 'xgb':
            preds, model_obj = train_xgboost(X_train, y_train, X_test)
        elif model_name == 'lstm':
            preds, model_obj = train_lstm(X_train, y_train, X_test, epochs=30)
        else:
            raise ValueError('Modele non reconnu')

        preds = np.array(preds).reshape(-1)
# aligner tailles: si mismatch, tronquer et avertir
        if len(preds) != len(y_test):
            min_len = min(len(preds), len(y_test))
            print(f"Model {model_name} retourne {len(preds)} preds parceque y_test has {len(y_test)}; tronquer em {min_len}.")
            preds = preds[:min_len]
            y_test_eff = y_test[:min_len]
            index_eff = test.index[:min_len]
        else:
            y_test_eff = y_test
            index_eff = test.index[:len(preds)]

        metrics = evaluation(y_test_eff, preds)
        all_metrics.append(metrics)
        derniere_preds = (index_eff, preds, y_test_eff)
# calculer moyenne des metrics
    moy = {k: np.mean([m[k] for m in all_metrics]) for k in all_metrics[0].keys()}
    return all_metrics, moy, derniere_preds
```

```python
def choix_du_meilleur_model(metric_dict, critere='RMSE'):
    if not metric_dict:
        return None, None
    # s√©curit√©: lowercase keys handling
    critere = critere if isinstance(critere, str) else 'RMSE'
    # filtrer les mod√®les qui contiennent la m√©trique demand√©e
    valid = {m: metric_dict[m] for m in metric_dict if isinstance(metric_dict[m], dict) and critere in metric_dict[m]}
    if not valid:
        #retourner premier mod√®le
        first = next(iter(metric_dict))
        return first, metric_dict[first]
    meill_model = None
    meill_val = None
    for m, vals in valid.items():
        try:
            v = float(vals[critere])
        except Exception:
            continue
        if meill_val is None or v < meill_val:
            meill_val = v
            meill_model = m
    return meill_model, meill_val
```

```python
def run_pipeline(csv_path=None, date_col=None, cible=None, exog_cols=None, resample_rule=None,apply_decomposition=True):
   
    print('1) Chargement du CSV')
    df_raw = chargement_dataset(csv_path=csv_path)

    print('\n2) Aggregation journaliere (moyenne)')
    df_compiler = compilation_donnees(df_raw,cible=cible,date_col=date_col, colonnes=exog_cols, rule=resample_rule)
    print('dmension apres compilation:', df_compiler.shape)
# Visualiser la serie brut et la serie agr√©g√©e
    plt.figure(figsize=(12,4))
    if cible in df_compiler.columns:
        plt.plot(df_compiler[cible], label=f'{cible} compilation moyenne')
    else:
# si la cible n'est pas pr√©sente, tracer la premi√®re colonne pour debug
        first_col = df_compiler.columns[0] if len(df_compiler.columns) > 0 else None
        if first_col is not None:
            plt.plot(df_compiler[first_col], label=f'{first_col} compilation moyenne (cible absente)')
    plt.title('Serie compilee (moyenne journaliere)')
    plt.legend()
    plt.savefig(os.path.join(OUT_DIR, 'series_daily.png'))
    plt.show()

    print('\n3) Nettoyage: valeurs ab et manq')
    df_clean =Valeurs_manquantes_et_aberantes(df_compiler, method='interpolate',seuil_manquant=0.7 ,tolerance=0.2)
    
    print('\n4Ô∏è‚É£ D√©composition de la s√©rie temporelle')
    df_stationaire = df_clean.copy()
    if apply_decomposition:
        try:
            df_decomp, info = decomposition_serie_temporelle(df_clean, cible=cible, model='auto', plot=True, out_dir=OUT_DIR)
            df_stationaire[cible] = df_decomp['cible_decomp']
            print(" S√©rie d√©compos√©e et rendue stationnaire.")
        except Exception as e:
            print(f"D√©composition √©chou√©e ({e}) ‚Äî utilisation de la s√©rie originale.")
    else:
        print(" √©composition d√©sactiv√©e (s√©rie originale conserv√©e")
    try:
        models_run = ['ridge', 'lasso', 'ar', 'rf']
        models_run.append('sarimax')
        if XGBOOST_AVAILABLE:
            models_run.append('xgb')
        if TF_AVAILABLE:
            models_run.append('lstm')
    except Exception:
         models_run = ['ridge', 'lasso', 'ar', 'rf']
         models_run.append('sarimax')
    liste_des_metric = {}
    preds_for_plot = {}
    for mname in models_run:
        print(f'\nMod√®le : {mname} ---')
# Choisir le bon dataset selon le mod√®le
        if mname in ['ar', 'sarimax']:
            print("Utilisation de la s√©rie d√©compos√©e/stationnaire")
            df_used = df_stationaire.copy()  # <-- s√©rie d√©compos√©e
        else:
            print("Utilisation de la s√©rie originale")
            df_used = df_clean.copy()  # <-- s√©rie originale
    
        print('\n5 Cr√©ation des features')
        df_feat = creation_de_features(df_used, cible=cible, lags=[1,2,3,7,14], rolling_windows=[7,14,30])
        print('Shape features:', df_feat.shape)
    
        print('\n6 D√©coupage train/test')
        train_df, test_df = Decoupage_train_test_tempo(df_feat, test_size=0.2)
        print('Train shape:', train_df.shape, 'Test shape:', test_df.shape)

        feature_cols = [c for c in df_feat.columns if c not in selection_colonne and c != 'counts' and c != cible]
    
        print('\n7 Normalisation')
        scaler, train_s, test_s = Normalisatiion_donnees(train_df, test_df, feature_cols)
        try:
            if scaler is not None:
                joblib.dump(scaler, os.path.join(OUT_DIR, f'scaler_{mname}.joblib'))
        except Exception:
            pass 

        print('\n6 Entrainement et evaluation via walk-forward')
        try:
            # concat√©ner train/test normalis√©s pour validation
            df_concat = pd.concat([train_s, test_s]).sort_index()
            all_metrics, moy_m, derniere_preds = walk_forward_validation(df_concat, feature_cols, cible=cible, model_name=mname, n_splits=5, horizon=HORIZON)
            liste_des_metric[mname] = moy_m
            preds_for_plot[mname] = derniere_preds
        except Exception as e:
            print(f"Erreur pendant walk-forward pour {mname}: {type(e).__name__} {e}")
            continue
   
    if len(preds_for_plot) > 0:
        plt.figure(figsize=(12,5))
        for m, (idx, preds, y_true) in preds_for_plot.items():
            try:
                plt.plot(idx, preds, '--', label=f'{m} pred')
            except Exception:
                continue
        sample_idx = None
        try:
            sample_idx = list(preds_for_plot.values())[0][0]
        except Exception:
            sample_idx = None
        if sample_idx is not None:
            try:
                # trouver vrai values dans df_feat (aligner par index)
                if cible in df_feat.columns:
                    true_vals = df_feat.loc[sample_idx, cible]
                    plt.plot(sample_idx, true_vals, label='True', color='black')
            except Exception:
                pass
        plt.legend()
        plt.title('Comparaison predictions (derniere fenetre)')
        plt.savefig(os.path.join(OUT_DIR, 'derniere_preds.png'))
        plt.show()

    # Sauvegarde des metriques
    if len(liste_des_metric) > 0:
        metrics_df = pd.DataFrame(liste_des_metric).T
        metrics_df.to_csv(os.path.join(OUT_DIR, 'resume_models_metrics.csv'))
        print('\n resume des metrics sauvegarder', os.path.join(OUT_DIR, 'resume_models_metrics.csv'))

    
    meill_model, meill_val = choix_du_meilleur_model(liste_des_metric, critere='RMSE')
    if meill_model is not None:
        print(f"\nMeilleur mod√®le selon RMSE: {meill_model} (RMSE={meill_val})")
      
        try:
            with open(os.path.join(OUT_DIR, 'meilleur_model.txt'), 'w', encoding='utf-8') as f:
                f.write(f"{meill_model}\n{meill_val}\n")
        except Exception:
            pass

  
    if 'rf' in liste_des_metric:
        # entrainer RF final sur tout le training set
        rf_model = RandomForestRegressor(n_estimators=200, random_state=0)
        try:
            if len(feature_cols) > 0:
                rf_model.fit(pd.concat([train_s[feature_cols], test_s[feature_cols]]), pd.concat([train_s[cible], test_s[cible]]))
                joblib.dump(rf_model, os.path.join(OUT_DIR, 'random_forest_final.joblib'))
        except Exception as e:
            print(f"[run_pipeline] impossible de sauvegarder RF final: {e}")
    print('\nGeneration d\'un notebook explicatif (.ipynb)')
    nb = new_notebook()
    md_intro = """
    # Pipeline serie temporelle Notebook
    Ce notebook a ete genere automatiquement 
    Il reprend les etapes: chargement, pre-traitement, features, entrainement, validation et visualisations.
    """
    nb.cells.append(new_markdown_cell(md_intro))
    
    code_setup = f"""
    # Imports essentiels
    import pandas as pd
    import numpy as np
    import matplotlib.pyplot as plt
    
    # Charger le fichier agrege
    csv_path = '{csv_path}'
    print('Modifier csv_path si necessaire')
    """
    nb.cells.append(new_code_cell(code_setup))
    
    # Ajouter une cellule markdown et code qui charge le fichier et affiche un preview
    nb.cells.append(new_markdown_cell('## Chargement et preview du dataset'))
    code_load = f"""
    df = pd.read_csv('{csv_path}', parse_dates=['{date_col}'])
    df.head()
    """
    nb.cells.append(new_code_cell(code_load))
    
    # Sauvegarder le notebook sur disque
    nb_path = os.path.join(OUT_DIR, 'air_quality_pipeline_output.ipynb')
    try:
        with open(nb_path, 'w', encoding='utf-8') as f:
            nbformat.write(nb, f)
        print('Notebook saved to', nb_path)
    except Exception as e:
        print(f"Impossible de sauvegarder le notebook: {e}")
    
    print('\nPipeline termin√©. Verifier le dossier', OUT_DIR, 'pour sorties (figures, models, notebook).')
    
    return {
        'metrics': liste_des_metric,
        'out_dir': OUT_DIR,
        'notebook': nb_path,
        'meilleur_model': meill_model,
        'meilleur_value': meill_val
    }
if __name__ == '__main__':
# V√©rifier que le CSV existe
    if not os.path.exists(CSV_PATH):
        print('ERREUR: le fichier CSV n\'a pas √©t√© trouv√©:', CSV_PATH)
        print('Place ton CSV dans le m√™me dossier ou modifie la variable CSV_PATH en haut du script.')
        sys.exit(1)
 
results = run_pipeline(csv_path=CSV_PATH, date_col=index, cible=Variable_cible,  exog_cols=selection_colonne, resample_rule=Reechantillonage)
print('\nResultats resume:')
print(results['metrics'])
```
